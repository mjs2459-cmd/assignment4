```python
# Part III - Question 1

def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

```python
kelvin_to_celsius(273.15)
celsius_to_kelvin(0)
```

```python
# Part III - Question 2

def to_fahrenheit(temp, unit="celsius"):
    if unit == "celsius":
        return (temp * 9/5) + 32
    elif unit == "kelvin":
        c = kelvin_to_celsius(temp)
        return (c * 9/5) + 32
```

```python
to_fahrenheit(0)
to_fahrenheit(273.15, unit="kelvin")
```

```python
to_fahrenheit(100)
to_fahrenheit(310, unit="kelvin")
```

```python
# Part III - Question 4

def from_fahrenheit(temp_f, output="celsius"):
    c = (temp_f - 32) * 5/9
    
    if output == "celsius":
        return c
    elif output == "kelvin":
        return celsius_to_kelvin(c)
```

```python
from_fahrenheit(32)
from_fahrenheit(32, output="kelvin")
```

```python
# Part III - Question 5

def height_to_meters(feet, inches):
    return feet * 0.3048 + inches * 0.0254
```

```python
height_to_meters(5, 7)
```

```python
# Part III - Question 6

def meters_to_height(meters):
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches % 12
    return feet, inches
```

```python
meters_to_height(1.70)
```

```python
# Part III - Question 7

heights = [1.5, 1.7, 1.85]

for h in heights:
    f, i = meters_to_height(h)
    back_to_meters = height_to_meters(f, i)
    print("Original:", h)
    print("Round trip:", back_to_meters)
    print()
```
